```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import scipy.stats as st
```

```python
# Show path to csv files
original_data = "Resources/completData.csv"

# Read csv files through Pandas
data = pd.read_csv(original_data)
```

```python
data.head(200)
```

```python
# Create a group based on the values in the 'region' column
region_group = data.groupby('Region')

# Count how many times each region appears in our group
count_regions = region_group['Region'].count()

count_regions
```

```python
# Create a bar chart based off of the Regions
count_chart = count_regions.plot(kind='bar', title="Region Bias")

# Set the xlabel and ylabel using class methods
count_chart.set_xlabel("Regions")
count_chart.set_ylabel("Number of Countries in Region")


plt.show()
plt.tight_layout()
```

```python
# Group data with Region as index
complet_data_Summary = data.groupby("Region")
complet_data_Summary.mean()
```

```python
list(data.columns)
```

```python
# Summary of Generocity per year
Trust_df = data[["Country", "Region", "Trust (Government Corruption) 2015", "Trust (Government Corruption) 2016", "Trust (Government Corruption) 2017", "Trust (Government Corruption) 2018", "Trust (Government Corruption) 2019"]]
Trust_df = Trust_df.groupby("Region")
Trust_df.mean()

#Find the averages of the data
GOV_df = Trust_df.mean().reset_index()
GOV_df["Gov Trust Average"] = GOV_df.mean(axis=1)
GOV_df

# Filter the DataFrame down only to those columns to chart
GOV_plot = GOV_df[["Region", "Gov Trust Average"]]

# Set the index to be "Region" so they will be used as labels
GOV_plot = GOV_plot.set_index("Region")
GOV_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
GOV_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average Government Trust Score Per Region")

plt.show()
plt.tight_layout()
```

```python
# Summary of Generocity per year
Generocity_df = data[["Country", "Region", "Generosity 2015", "Generosity 2016", "Generosity 2017", "Generosity 2018", "Generosity 2019"]]
Generocity_df = Generocity_df.groupby("Region")
Generocity_df.mean()

#Find the averages of the data
Gen_df = Generocity_df.mean().reset_index()
Gen_df["Generocity Average"] = Gen_df.mean(axis=1)
Gen_df

# Filter the DataFrame down only to those columns to chart
Gen_plot = Gen_df[["Region", "Generocity Average"]]

# Set the index to be "Region" so they will be used as labels
Gen_plot = Gen_plot.set_index("Region")
Gen_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
Gen_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average Generocity Score Per Region")

plt.show()
plt.tight_layout()
```

```python
# Summary of Health per year
Health_df = data[["Country", "Region", "Health (Life Expectancy) 2015","Health (Life Expectancy) 2016", "Health (Life Expectancy) 2017", "Health (Life Expectancy) 2018", "Health (Life Expectancy) 2019"]]
Health_df = Health_df.groupby("Region")
Health_df.mean()

#Find the averages of the data
Heal_df = Health_df.mean().reset_index()
Heal_df["Health Average"] = Heal_df.mean(axis=1)
Heal_df

# Filter the DataFrame down only to those columns to chart
Health_plot = Heal_df[["Region", "Health Average"]]

# Set the index to be "Region" so they will be used as labels
Health_plot = Health_plot.set_index("Region")
Health_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
Health_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average Life Expectancy Per Region")

plt.show()
plt.tight_layout()
```

```python
# Summary of Family per year
Family_df = data[["Country", "Region", "Family 2015", "Family 2016", "Family 2017", "Family 2018", "Family 2019"]]
Family_df = Family_df.groupby("Region")
Family_df.mean()

#Find the averages of the data
Fam_df = Family_df.mean().reset_index()
Fam_df["Family Average"] = Fam_df.mean(axis=1)
Fam_df

# Filter the DataFrame down only to those columns to chart
Family_plot = Fam_df[["Region", "Family Average"]]

# Set the index to be "Region" so they will be used as labels
Family_plot = Family_plot.set_index("Region")
Family_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
Family_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average GDP Per Region")

plt.show()
plt.tight_layout()
```

```python
#Summary of Freedom score
Freedom_df = data[["Country", "Region", "Freedom 2015", "Freedom 2016", "Freedom 2017", "Freedom 2018", "Freedom 2019"]]
Freedom_df = Freedom_df.groupby("Region")
Freedom_df.mean()

#Find the averages of the data
Free_df = Freedom_df.mean().reset_index()
Free_df["Freedom Average"] = Free_df.mean(axis=1)
Free_df

#sort data in ascending order based on Average
Free_df.sort_values(by=['Freedom Average'], inplace=True, ascending=False)
Free_df

# Filter the DataFrame down only to those columns to chart
Freedom_plot = Free_df[["Region", "Freedom Average"]]

# Set the index to be "Region" so they will be used as labels
Freedom_plot = Freedom_plot.set_index("Region")
Freedom_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
Freedom_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average Freedom Score Per Region")

plt.show()
plt.tight_layout()
```

```python
# Summary of Happiness per year
Happiness_df = data[["Country", "Region", "Happiness Rank 2015", "Happiness Rank 2016", "Happiness Rank 2017", "Happiness Rank 2018", "Happiness Rank 2019"]]
Happiness_df = Happiness_df.groupby("Region")
Happiness_df.mean()

#Find the averages of the data
Happ_df = Happiness_df.mean().reset_index()
Happ_df["Happiness Average"] = Happ_df.mean(axis=1)
Happ_df

# Filter the DataFrame down only to those columns to chart
Happiness_plot = Happ_df[["Region", "Happiness Average"]]

# Set the index to be "Region" so they will be used as labels
Happiness_plot = Happiness_plot.set_index("Region")
Happiness_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
Happiness_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average Happiness Score Per Region")

plt.show()
plt.tight_layout()
```

```python
# Summary of Happiness per year
Economy_df = data[["Country", "Region", "Economy (GDP per Capita) 2015","Economy (GDP per Capita) 2016", "Economy (GDP per Capita) 2017", "Economy (GDP per Capita) 2018", "Economy (GDP per Capita) 2019" ]]
Economy_df = Economy_df.groupby("Region")
Economy_df.mean()

#Find the averages of the data
Econ_df = Happiness_df.mean().reset_index()
Econ_df["Economy GDP Average"] = Econ_df.mean(axis=1)
Econ_df

# Filter the DataFrame down only to those columns to chart
Economy_plot = Econ_df[["Region", "Economy GDP Average"]]

# Set the index to be "Region" so they will be used as labels
Economy_plot = Economy_plot.set_index("Region")
Economy_plot.head()
Economy_plot.head()
```

```python
# Use DataFrame.plot() in order to create a bar chart of the data
Economy_plot.plot(kind="bar", figsize=(5,3))

# Set a title for the chart
plt.title("Average GDP Per Region")

plt.show()
plt.tight_layout()
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(Economy_plot))
tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.bar(x_axis, Economy_plot["Average"], color='r', alpha=1, align="center")
plt.xticks(tick_locations, Economy_plot["Region"], rotation="vertical")
plt.tight_layout()
```

```python
#Merge all average columns together 
# Merging Data on Region
Comparison_data = pd.merge(GOV_plot, Gen_plot, on = ["Region"], how = "outer")
Comparison_data = pd.merge(Comparison_data, Health_plot, on = ["Region"], how = "outer")
Comparison_data = pd.merge(Comparison_data, Family_plot, on = ["Region"], how = "outer")
Comparison_data = pd.merge(Comparison_data, Economy_plot, on = ["Region"], how = "outer")
Comparison_data = pd.merge(Comparison_data, Freedom_plot, on = ["Region"], how = "outer")
Comparison_data = pd.merge(Comparison_data, Happiness_plot, on = ["Region"], how = "outer")
Comparison_data
```

```python
# Using .rename(columns={}) in order to rename columns
renamed_df = Comparison_data.rename(columns={"Average":"Economic GDP Average"})
renamed_df
renamed_df.to_csv("Resources/Region_output.csv")
```

```python
# The next example will compute the Pearson correlation coefficient between "IEconomic GDP Average" and "Freedom Average"
GDP = renamed_df.iloc[:,4]
Free = renamed_df.iloc[:,5]
correlation = st.pearsonr(GDP,Free)
print(f"The correlation between both factors is {round(correlation[0],2)}")
```

```python
# Compare the calcualted Pearson's r to the plots
plt.scatter(GDP,Free)
plt.xlabel('Region GDP Per Capita')
plt.ylabel('Average Freedom Score Per Region Per Year (L)')
print(f"The correlation between both factors is {round(correlation[0],2)}")
plt.show()
```
